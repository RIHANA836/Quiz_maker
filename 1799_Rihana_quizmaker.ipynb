```python
from tkinter import *
from tkinter import ttk

def next_tab(tab_control):
    tab_control.select(tab2)


def submit_answers(selected1, selected2, selected3, selected4, selected5, selected6, selected7, selected8):
    correct_answers = {
        1: "Hedwig",     
        2: "Gryffindor", 
        3: "Expelliarmus", 
        4: "Severus Snape", 
        5: "Seeker",     
        6: "Albus Dumbledore",  
        7: "The Philosopher's Stone", 
        8: "Scabbers" 
    }

    user_answers = {
        1: selected1.get(),
        2: selected2.get(),
        3: selected3.get(),
        4: selected4.get(),
        5: selected5.get(),
        6: selected6.get(),
        7: selected7.get(),
        8: selected8.get()
    }

    score = 0
    for i in range(1, 9):
        if user_answers[i] == correct_answers[i]:
            score += 1
    results_text = f"Your Score: {score}/8\n\n"
    
    lb.config(text=results_text)
    tab_control.select(tab3)     


window = Tk()
window.title("Quiz")
window.geometry("600x500")

lbl = Label(window, text="Welcome to the Quiz", font=("Helvetica", 12))
lbl.pack(pady=10)
lbl1 = Label(window, text="Uncover Your Hogwarts Knowledge.\nTest your Harry Potter knowledge with this fun quiz!", font=("Arial", 10))
lbl1.pack(pady=10)

tab_control = ttk.Notebook(window)
tab_control.pack(expand=1, fill='both')

tab1 = ttk.Frame(tab_control)
tab_control.add(tab1, text='Questions 1-4')

selected1 = StringVar(value="")  
selected2 = StringVar(value="")  
selected3 = StringVar(value="")  
selected4 = StringVar(value="")  

q1 = Label(tab1, text="1. What is the name of Harry Potter's pet owl?")
q1.pack(anchor=W)
rad1 = Radiobutton(tab1, text='A) Fawkes', value='Fawkes', variable=selected1)
rad2 = Radiobutton(tab1, text='B) Hedwig', value='Hedwig', variable=selected1)
rad1.pack(anchor=W)
rad2.pack(anchor=W)

q2 = Label(tab1, text="2. Which house at Hogwarts does Harry belong to?")
q2.pack(anchor=W)
rad3 = Radiobutton(tab1, text='A) Gryffindor', value='Gryffindor', variable=selected2)
rad4 = Radiobutton(tab1, text='B) Hufflepuff', value='Hufflepuff', variable=selected2)
rad3.pack(anchor=W)
rad4.pack(anchor=W)

q3 = Label(tab1, text="3. What spell is used to disarm an opponent?")
q3.pack(anchor=W)
rad5 = Radiobutton(tab1, text='A) Stupefy', value='Stupefy', variable=selected3)
rad6 = Radiobutton(tab1, text='B) Expelliarmus', value='Expelliarmus', variable=selected3)
rad5.pack(anchor=W)
rad6.pack(anchor=W)

q4 = Label(tab1, text="4. Who is the Half-Blood Prince?")
q4.pack(anchor=W)
rad7 = Radiobutton(tab1, text='A) Draco Malfoy', value='Draco Malfoy', variable=selected4)
rad8 = Radiobutton(tab1, text='B) Severus Snape', value='Severus Snape', variable=selected4)
rad7.pack(anchor=W)
rad8.pack(anchor=W)

Button(tab1, text="Next", command=lambda: next_tab(tab_control)).pack(pady=10)

tab2 = ttk.Frame(tab_control)
tab_control.add(tab2, text='Questions 5-8')

selected5 = StringVar(value="")  
selected6 = StringVar(value="")  
selected7 = StringVar(value="")  
selected8 = StringVar(value="")  

q5 = Label(tab2, text="5. What position does Harry play on the Gryffindor Quidditch team?")
q5.pack(anchor=W)
rad9 = Radiobutton(tab2, text='A) Keeper', value='Keeper', variable=selected5)
rad10 = Radiobutton(tab2, text='B) Seeker', value='Seeker', variable=selected5)
rad9.pack(anchor=W)
rad10.pack(anchor=W)

q6 = Label(tab2, text="6. Who was the headmaster of Hogwarts during Harry's first year?")
q6.pack(anchor=W)
rad11 = Radiobutton(tab2, text='A) Albus Dumbledore', value='Albus Dumbledore', variable=selected6)
rad12 = Radiobutton(tab2, text='B) Severus Snape', value='Severus Snape', variable=selected6)
rad11.pack(anchor=W)
rad12.pack(anchor=W)

q7 = Label(tab2, text="7. Which potion is known as the 'Elixir of Life'?")
q7.pack(anchor=W)
rad13 = Radiobutton(tab2, text='A) Polyjuice Potion', value='Polyjuice Potion', variable=selected7)
rad14 = Radiobutton(tab2, text='B) The Philosopher\'s Stone', value='The Philosopher\'s Stone', variable=selected7)
rad13.pack(anchor=W)
rad14.pack(anchor=W)

q8 = Label(tab2, text="8. What is the name of Ron Weasley's pet rat?")
q8.pack(anchor=W)
rad15 = Radiobutton(tab2, text='A) Scabbers', value='Scabbers', variable=selected8)
rad16 = Radiobutton(tab2, text='B) Crookshanks', value='Crookshanks', variable=selected8)
rad15.pack(anchor=W)
rad16.pack(anchor=W)

Button(tab2, text="Submit",command=lambda: submit_answers(selected1, selected2, selected3, selected4, selected5, selected6, selected7, selected8)).pack(pady=10)
      

tab3 = ttk.Frame(tab_control)
tab_control.add(tab3, text='Results')

lb = Label(tab3, text="", font=("Arial", 12))
lb.pack(pady=20)


window.mainloop()
```
